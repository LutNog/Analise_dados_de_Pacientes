# Análise Exploratória e Descritiva de Dados de Pacientes.

```python
import numpy as nd
import pandas as pd
import statistics as sts
import seaborn as srn
import scipy.stats as stats
```

```python
dados = pd.read_csv('dados_pacientes.csv')
dados
```

```python
dados.shape
```

```python
# Explorando dados categóricos
# tipo sanguineo
agrupar = dados.groupby(['tipo_sanguineo']).size()
agrupar
```

```python
sts.mode(dados['tipo_sanguineo'])
```

```python
tipo_sanguineo = dados.iloc[:,4]
srn.histplot(tipo_sanguineo)
```

```python
# estado civil

agrupar = dados.groupby(['estado_civil']).size()
agrupar
```

```python
sts.mode(dados['estado_civil'])
```

```python
estado_civil = dados.iloc[:,5]
srn.histplot(estado_civil)
```

```python
# Explorando dados numéricos
# idade

dados['idade'].describe()
```

```python
# Variância
sts.variance(dados['idade'])
```

```python
# Moda
sts.mode(dados['idade'])
```

```python
# Testando a distribuição dos dados.
stats.shapiro(dados['idade'])
```

```python
# Verificando distribuição dos dados

srn.distplot(dados["idade"]).set_title("Idade")
```

```python
# Verificando Outliers

srn.boxplot(dados['idade']).set_title('Idade')
```

```python
# Explorando dados numéricos
# altura

dados["altura"].describe()
```

```python
sts.variance(dados['altura'])
```

```python
sts.mode(dados['altura'])
```

```python
# Testando a distribuição dos dados.
stats.shapiro(dados['altura'])
```

```python
srn.distplot(dados['altura']).set_title('Altura')
```

```python
srn.boxplot(dados['altura']).set_title('Altura')
```

```python
dados["peso"].describe()
```

```python
sts.variance(dados['peso'])
```

```python
sts.mode(dados['peso'])
```

```python
# Testando a distribuição dos dados.
stats.shapiro(dados['peso'])
```

```python
srn.distplot(dados['peso']).set_title("Peso")
```

```python
srn.boxplot(dados['peso']).set_title("Peso")
```
